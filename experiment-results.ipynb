```python
import pandas as pd
import json
from pathlib import Path
import os
```

```python
filepath = Path('experiments') / 't5-small_falcon2-default_annotation-k1' / 'validations' / 'val_final.json'

with open(filepath) as f:
  val_dict = json.load(f)

correct_count = 0
total_count = 0
for entry in val_dict:
  if entry['Gol2'] == entry['Gen2']:
    correct_count += 1
  total_count += 1

print(correct_count / total_count)
```
