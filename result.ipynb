# Results

This notebook is to view the dev_results when evaluating the model.

```python
import json
import os
from pathlib import Path
import pandas as pd
import re

Path.cwd()
```

```python
path = Path("baseline") / "lcquad2" / "preprocessed_data.csv"
```

```python
model_dict = {
    "small": "text2sparql-t5-small-lcquad2",
    "base": "text2sparql-t5-base-lcquad2",
}

model_key = "small"

model_path = Path().cwd().joinpath('models').joinpath(model_dict[model_key])
model_evals_dir = model_path.joinpath("evaluations")
model_evals_dir = Path(r"baseline/lcquad2")

def retrieve_latest(path, regex):
    latest = 0
    res = None
    for res_path in path.iterdir():
        path_search = re.search(regex, res_path.name, re.IGNORECASE)
        if path_search and int(path_search.group(1)) > latest:
            latest = int(path_search.group(1))
            res = res_path

        print(res)
    
    return res

latest_eval_path = retrieve_latest(model_evals_dir, r"dev_result([0-9]+)")
```

```python
with open(latest_eval_path) as f:
    latest_eval_json = json.load(f)
df_raw = pd.DataFrame.from_records(latest_eval_json)
df_raw.head()
```

```python
df_raw= df_raw.convert_dtypes()
df_raw.gold = df_raw.gold.str.replace(" ", "")
df_raw.generated = df_raw.generated.str.replace(" ", "")
df_raw
```

```python
df_true = df_raw[df_raw["gold"] == df_raw["generated"]]
df_false = df_raw[df_raw["gold"] != df_raw["generated"]]
assert len(df_raw) == len(df_true) + len(df_false)
```
